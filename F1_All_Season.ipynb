```python
#imports
import json
import pandas as pd
import numpy as np
import requests
```

```python
#dataframe list to store evry gathered dataframes from the json response.
appended_list=[]
for year in list(range(2019,2024)):
    url = 'https://ergast.com/api/f1/{}.json'
    r = requests.get(url.format(year))
    json1 = r.json()
    df=pd.json_normalize(json1)
    appended_list.append(df[['MRData.RaceTable.Races']])
```

```python
#Converting list of lists of dictionaries
new_list=[]
final_list=[]
for element in appended_list:
    listDicts=element['MRData.RaceTable.Races']
    for i in range(0,len(listDicts),):
        new_list.append(list(listDicts[i]))
for i in range(0,len(new_list),):
    for j in range(0,len(new_list[i]),):
        final_list.append(new_list[i][j])
final_df=pd.DataFrame(final_list)
final_df=final_df.rename(columns={'url':'Race_url','date':'Race_date','time':'Race_time'})
print(final_df)
```

```python
#flattening the circuit column dictionaries in this cell
final_df=pd.concat([final_df.drop('Circuit', axis=1), pd.json_normalize(final_df['Circuit'])], axis=1)
final_df=final_df.rename(columns={'url':'Circuit_url'})
final_df=pd.concat([final_df.drop('FirstPractice', axis=1), pd.json_normalize(final_df['FirstPractice'])], axis=1)
final_df=final_df.rename(columns={'date':'FP_date','time':'FP_time'})
final_df=pd.concat([final_df.drop('SecondPractice', axis=1), pd.json_normalize(final_df['SecondPractice'])], axis=1)
final_df=final_df.rename(columns={'date':'SP_date','time':'SP_time'})
final_df=pd.concat([final_df.drop('ThirdPractice', axis=1), pd.json_normalize(final_df['ThirdPractice'])], axis=1)
final_df=final_df.rename(columns={'date':'TP_date','time':'TP_time'})
final_df=pd.concat([final_df.drop('Qualifying', axis=1), pd.json_normalize(final_df['Qualifying'])], axis=1)
final_df=final_df.rename(columns={'date':'Quali_date','time':'Quali_time'})
final_df=pd.concat([final_df.drop('Sprint', axis=1), pd.json_normalize(final_df['Sprint'])], axis=1)
final_df=final_df.rename(columns={'date':'Sprint_date','time':'Sprint_time'})
print(final_df)
```

```python
#writing to excel or generating the excel code
final_df.to_csv("AllSeasons.csv")
```
